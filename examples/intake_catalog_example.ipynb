### Intake Catalog Example

```python
import doralite as dl
import xarray as xr
import intake_esm
```

```python
import warnings
# Suppress ONLY FutureWarnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

#### Load the catalog from the Dora server

```python
cat = dl.catalog("odiv-496")
cat
```

```python
cat.df
```

#### Query the catalog for sea surface temperature

```python
query = cat.search(variable_id=['tos'],realm="ocean_monthly")
query
```

```python
query.df
```

#### Get a list of paths and dmget files 

```python
paths = sorted(query.df['path'].tolist())
paths
```

```python
dl.dmget(paths)
```

#### Open the xarray dataset

```python
ds = xr.combine_by_coords(query.to_dataset_dict().values())
```

```python
ds
```

#### Use case example:  time average and make a plot

```python
import momlevel as ml
```

```python
# Generate annual mean climatology from monthly values
dsmean = ml.util.annual_average(ds).mean("time")
```

```python
# Up to this point, everything has been evaluated lazily
tos = dsmean.tos.load()
```

```python
dsmean.tos.plot()
```
